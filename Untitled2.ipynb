#  ANALYSING THE POWER CONSUMPTION OF MULTIPLE ACs IN A HOTEL IN    GURGAON

The first step in the process of data analytics is refining the data devoid of missing data and other abnormalities. This process is called "data cleaning" or "data wrangling" or "data munging".

As a part of the data wrangling process, I have done the following things with the given dataset.
1) Importing the dataset given in the form of a .csv file as a dataframe to the Jupyter notebook.
2) Indexing the dataset with the field "date".
3) Transforming the type of the date field from string to timestamp for further processing.
4) Checking for any abnormalities or holes in the data.

```python
import pandas as pd
```

```python
rawacdata=pd.read_csv("E:\AC_Data.csv",parse_dates=["Timestamp"], index_col="Timestamp")
rawacdata.tail()
```

## The below table gives the statiscally important information regarding the dataset containing the 'minute-wise power consumption' data of different ACs

```python
rawacdata.describe()
```

## Hourly power consumption is a relevant and more useful data when compared to minute-wise power consumption. Hence the data frequency has been transformed from minutes to hours. 

```python
hourly_acdata=rawacdata.resample('H').mean()
hourly_acdata
```

## The following table displays the statistically important information of the hourly power consumption dataframe: 

```python
hourly_acdata.describe()
```

## From the above statistical data, we can infer that the mean power consumption of "AC_18" is higher than any other AC for the given time period. Also, the mean power consumption of "AC_5 is the least among all the ACs. Hence,

#  AC_18 is the highest power consuming AC.
#  AC_5 is the least power consuming AC.



## In order to compare the power consumption data with the temperature of Gurgaon, I have downloaded a dataset from the webpage "kaggle".

## The dataset span was for an enormous duration and there were several weather parameters listed. I filtered out the necessary fields for the required time span and created a data frame.  

```python
tempdata = pd.read_csv("E:\gurgaon.csv",parse_dates=["Timestamp"], index_col="Timestamp",usecols=['Timestamp','tempC'])
tempdata.head()
```

```python
tempdata.index
```

```python
type(tempdata.index[0])
```

```python
hourly_tempdata=tempdata.resample('H').mean()
hourly_tempdata.head()
```

```python
hourly_tempdata_req=hourly_tempdata['2019-01-08 00:00:00':'2019-12-09 23:00:00']
hourly_tempdata_req
```

## After cleaning the temperature dataframe, it is merged with the power consumption dataframe.

```python
process_data = pd.merge(rawacdata,hourly_tempdata_req , on="Timestamp")
process_data
```

```python
process_data.describe()
```

## A plot is created to evaluate  patterns in the dataset and employ the decoded patterns for the purpose of predictive analytics. 

```python
import matplotlib.pyplot as plt
process_data.plot(xlabel='Timestamp',ylabel='tempC',kind="bar")
plt.show()
```
